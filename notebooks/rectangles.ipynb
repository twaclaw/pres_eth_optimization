```julia
using Luxor
using LaTeXStrings
```

```julia
using Plots
```

```julia
function rect_dims(x; L=10)
    ℓ = x*L
    w = (L-2ℓ)/2
    return ℓ, w, round((x - 2x^2)/2,digits=4), round((2x)/(1-2x), digits=4)
end

function new_plot(;L=10)
    plot(ylabel = L"\ell", xlabel = L"w",
        xlims = (0,L), ylims = (0, L), opacity=.3,
        xticks=false, yticks=false, aspect_ratio=:equal, fmt=:svg)
end 
```

```julia
rectangle(w, h, x, y) = Shape(x .+ [0,w,w,0], y .+ [0,0,h,h])

```

```julia
p = new_plot(;L=5)
for i in [0.1, 0.25, 0.3, 0.4]
    ℓ,w,area,x = rect_dims(i)
    plot!(p, rectangle(ℓ,w, 0,0), opacity=0.4, label=latexstring("\$\\ell=$(x)w \\quad A=$(area)L^2\$"))
end
```

```julia
plot!(p)
```
