```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

# sns.set()
# sns.set_theme()
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
# plt.style.use('dark_background')
```

```python
def f(x, L=4):
    return -x**2 + x*L/2

x = np.linspace(-5, 5, 100)
y = f(x)
plt.xticks([], [])
plt.yticks([], [])
plt.xlabel('$w$')
plt.ylabel('$f(w)$')
plt.plot(x, y)
plt.savefig("../figures/x2_1.svg")
```

```python


def df(x, L=4):
	return -2*x + L/2

def tangent(x, x1, y1):
    return df(x1)*(x - x1) + y1

def plot_line(x_, y_):
    xrange = np.linspace(x_-1, x_+1, 10)
    plt.plot(xrange, tangent(xrange, x_, y_), 'C1--', linewidth=1.5)
    plt.scatter(x_, y_, s=50, color='b')

x1, x2, x3 = -3, 1, 4
y1, y2, y3 = f(x1), f(x2), f(x3)


plt.xticks([], [])
plt.yticks([], [])
plt.xlabel('$w$')
plt.ylabel('$f(w)$')
plt.plot(x, y)

plot_line(x1, y1)
plot_line(x2, y2)
plot_line(x3, y3)

plt.xticks([1], ['$w^{*}=L/4$'])
plt.savefig("../figures/x2_2.svg")


```

```python
N = 20
x = np.linspace(0, N, 200)
y = x**2*(np.sin(x))
plt.ylim(-320, 320)
plt.xticks([], [])
plt.yticks([], [])
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.plot(x,y, '-')
```

```python
plt.xticks([], [])
plt.yticks([], [])
plt.ylim(-320, 320)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.plot(x,y, '-')
plt.plot([17.4], [-298], marker="o", markersize=10, color='m')
plt.xticks([17.4], ['$x^*$'])

plt.savefig("../figures/critical_points1.svg")
```

```python
plt.xticks([], [])
plt.yticks([], [])
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.ylim(-320, 320)
plt.plot(x,y, '-')
plt.plot([17.4], [-298], marker="o", markersize=10, color='m')
plt.plot(x,-y, '--')
plt.plot([17.4], [298], marker="o", markersize=10, color='m')
plt.xticks([17.4], ['$x^*$'])

plt.savefig("../figures/critical_points2.svg")
```

```python
plt.plot(x,y, '-')
plt.xticks([], [])
plt.yticks([], [])
plt.ylim(-320, 320)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.axvspan(5, 13, color='green', alpha=0.2)
plt.plot([11.2], [-117], marker="o", markersize=10, color='m')
plt.xticks([11.2], ['$x^*$'])

plt.savefig("../figures/critical_points3.svg")
```
